## 4. Extração e padronização de variáveis da URL
- Extrair coluna `state` da URL  
- Normalizar estados em letras maiúsculas  
- Extrair coluna `city` da URL  
- Remover colunas desnecessárias (`avatar`)  

```python
# Cria coluna 'id' a partir da URL (identificador único do restaurante)
ifood_2020['id'] = ifood_2020['url'].str.split("/").str[-2]
ifood_2021['id'] = ifood_2021['url'].str.split("/").str[-2]

print(f"Em 2020 teve um total de: {len(ifood_2020['id'])} vendas")
print("\n")
print(f"Em 2021 teve um total de: {len(ifood_2021['id'])} vendas")
```

```python
# Extrai 'state' a partir da URL
ifood_2020['state'] = ifood_2020['url'].str.split("/").str[4].str.split("-").str[-1]
ifood_2021['state'] = ifood_2021['url'].str.split("/").str[4].str.split("-").str[-1]


# Conta número de linhas e estados únicos em 2020
print(f"Dados de 2020, total de linhas: {len(ifood_2020['state'])}")
print(f"Total de {ifood_2020['state'].nunique()} Estados na base de dados de 2020.")

print("\n")

# Conta número de linhas e estados únicos em 2021
print(f"Dados de 2021, Total de linhas: {len(ifood_2021['state'])}")
print(f"Total de {ifood_2021['state'].nunique()} Estados na base de dados de 2021.")

# Padroniza 'state' em letras maiúsculas
ifood_2020['state'] = ifood_2020['state'].str.upper()
ifood_2021['state'] = ifood_2021['state'].str.upper()
```

```python
# Extrai 'city' a partir da URL
ifood_2020['city'] = ifood_2020['url'].str.split("/").str[4].str.split("-").str[0]
ifood_2021['city'] = ifood_2021['url'].str.split("/").str[4].str.split("-").str[0]

print(f"Total de linhas:{len(ifood_2020['city'])} ")
print("\n")
print(f"Total de linhas:{len(ifood_2021['city'])} ")

# Padroniza 'city' em formato título (primeira letra maiúscula)
ifood_2021['city'] = ifood_2021['city'].str.title()
ifood_2020['city'] = ifood_2020['city'].str.title()
```

```python
# Remove colunas irrelevantes
ifood_2020 = ifood_2020.drop(columns=['avatar'])
ifood_2021 = ifood_2021.drop(columns=['avatar'])
```

## 6. Comparativo de restaurantes entre 2020 e 2021
- Contar número de restaurantes únicos por ano (coluna `id`)  
- Calcular a diferença entre os anos  
- Plotar gráfico de barras comparando os dois períodos  


```python
# Conta restaurantes únicos em 2020 e 2021 (com base no 'id')
qtd_2020 = ifood_2020['id'].nunique()
qtd_2021 = ifood_2021['id'].nunique()

# Calcula diferença entre os dois anos
dif = qtd_2021 - qtd_2020
```

```python
anos = ["2020", "2021"]
valores = [qtd_2020, qtd_2021]
cores = ["#FF7F0A", "#6A5ACD"]
# plot
fig, ax = plt.subplots(figsize=(6, 5), constrained_layout=True)
ax.bar(anos, valores, color=cores, width=0.80, zorder=3)

ax.set_ylim(0, max(valores) * 1.25)

# diferença de restaurantes - topo
ax.annotate(f'Diferença de Restaurantes: {dif:,.0f}'.replace(',', '.'),
            xy=(0.5, 0.98), xycoords='axes fraction',
            ha='center', va='top', fontsize=12, fontweight='bold')

# valores de cada barra
for x, v in zip(anos, valores):
    ax.text(x, v * 0.10, f'{v:,.0f}\nRestaurantes'.replace(',', '.'),
            ha='center', va='center', fontsize=12, fontweight='medium', color='black')

# rótulos
ax.set_title('Comparativo de Restaurantes no iFood (2020 vs 2021)', fontsize=12)
ax.grid(True, linestyle="--", alpha=1)
ax.tick_params(axis='y', labelsize=12)
ax.tick_params(axis='x', labelsize=12)
plt.show()
```

## 7. Comparativo de restaurantes por estado (2020 vs 2021)
- Contar número de restaurantes únicos por estado em cada ano  
- Criar DataFrame comparativo entre 2020 e 2021

```python
# Conta restaurantes únicos por estado em 2020
rest_por_estado_2020 = ifood_2020.groupby("state", observed=True)["id"].nunique()
# Conta restaurantes únicos por estado em 2021
rest_por_estado_2021 = ifood_2021.groupby("state", observed=True)["id"].nunique()

# Cria DataFrame comparativo (2020 vs 2021)
df_estados = pd.DataFrame({
    "2020": rest_por_estado_2020,
    "2021": rest_por_estado_2021
}).fillna(0).astype(int)

# Ordenar (ex: pelo total em 2021)
df_estados = df_estados.sort_values("2021", ascending=False)
```

```python
# Gráfico comparativo (barras lado a lado)
ax = df_estados.plot(kind="barh", figsize=(10,12), width=0.8, zorder=3, color={"2020": "#FF7F0A", "2021":"#6A5ACD"})

# Exibe valores
for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=8, padding=2)
    
# Ajusta título e rótulos
plt.title("Restaurantes por estado - Comparação 2020 vs 2021", fontsize=14)
plt.xlabel("Quantidade de Restaurantes", fontsize=12)
plt.ylabel("Estado", fontsize=12)
plt.grid(True, linestyle="--", alpha=1)
plt.tight_layout()
plt.show()
```

## 8. Top categorias mais vendidas (2020 vs 2021)

## 9. Análise inicial do `rating`
- Verificar estatísticas descritivas da variável de avaliação  
- Comparar distribuição das notas por faixa de preço em cada ano  
- Calcular a média do `rating` em cada faixa de preço  


```python
# Estatísticas descritivas do rating (2020)
print(ifood_2020['rating'].describe())
```

```python
# Bases e cores para o gráfico
bases = {"2020": ifood_2020, "2021": ifood_2021}
cores = {"2020": "#FF7F0A", "2021": "#6A5ACD"}

# Configura gráfico em 2 linhas
fig, axes = plt.subplots(2, 1, figsize=(10, 7), constrained_layout=True)

# Loop para plotar cada ano
for i, (ano, df) in enumerate(bases.items()):
    # Conta restaurantes por faixa de preço
    counts = df['price_range'].value_counts()
    # Calcula média do rating por faixa de preço (considera apenas > 0)
    means = df.loc[df['rating'] > 0].groupby('price_range')['rating'].mean().reindex(counts.index)

    # Plota gráfico de barras
    ax = axes[i]
    bars = ax.bar(counts.index, counts.values, color=cores[ano])
    ax.set_title(f"Avaliação por faixa de preço - {ano}", fontsize=12)
    ax.set_xlabel("Faixa de Preço", fontsize=10)
    ax.set_ylabel("Quantidade", fontsize=10)
    ax.tick_params(axis='x', labelsize=9)
    ax.tick_params(axis='y', labelsize=9)
    ax.grid(True, axis="both", linestyle="--", linewidth=0.9)  
    ax.set_axisbelow(True) 
    
    # Adiciona anotação da média do rating em cada coluna
    for bar, m in zip(bars, means):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
                f"Média: {m:.2f}", ha='center', va='bottom', fontsize=9)
plt.show()
```

## 10. Rating médio por faixas de tempo de entrega (2020 vs 2021)
- Criar bins de `delivery_time`  
- Calcular média de `rating` por faixa (ignorando zeros)  
- Comparar 2020 vs 2021 em gráficos lado a lado  


```python
# Bins e rótulos do tempo de entrega (min)
bins   = [0, 20, 40, 60, 120]
labels = ["até 20min","21-40min","41-60min","61-120min"]

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True, constrained_layout=True)

# Cria coluna de faixa de tempo (2020)
ifood_2020["faixa_tempo"] = pd.cut(ifood_2020["delivery_time"], bins=bins, labels=labels)
# Calcula média de rating por faixa de tempo em 2020 (ignora rating=0)
media_2020 = (ifood_2020.groupby("faixa_tempo", observed=True)["rating"]
    .apply(lambda x: x.where(x > 0).mean()))

# Gráfico 2020
media_2020.plot(kind="bar", color="#FF7F0A", ax=axes[0])
axes[0].set_title("Média de rating por tempo de entrega (2020)", fontsize=12)
axes[0].set_xlabel("Faixa de tempo de entrega", fontsize=10)
axes[0].set_ylabel("Rating (1-5)", fontsize=10)
axes[0].tick_params(axis="x", labelrotation=0)
axes[0].set_ylim(1, 5)
axes[0].grid(True, axis="both", linestyle="--", linewidth=0.9)
axes[0].set_axisbelow(True)


# Cria coluna de faixa de tempo (2021)
ifood_2021["faixa_tempo"] = pd.cut(ifood_2021["delivery_time"], bins=bins, labels=labels)
# Calcula média de rating por faixa de tempo em 2021 (ignora rating=0)
media_2021 = (ifood_2021.groupby("faixa_tempo", observed=True)["rating"]
    .apply(lambda x: x.where(x > 0).mean()))

# Gráfico 2021
media_2021.plot(kind="bar", color="#6A5ACD", ax=axes[1])
axes[1].set_title("Média de rating por tempo de entrega (2021)", fontsize=12)
axes[1].set_xlabel("Faixa de tempo de entrega", fontsize=10)
axes[1].tick_params(axis="x", labelrotation=0)
axes[1].set_ylim(1, 5)
axes[1].grid(True, axis="both", linestyle="--", linewidth=0.9)
axes[1].set_axisbelow(True)


plt.show()
```

## 11. Mediana do rating por faixas de tempo de entrega
- Calcular mediana do `rating` por faixa de tempo (2020 e 2021)  
- Usar mediana ao invés da média para maior robustez (menos sensível a outliers), assim confirmamos se há mudança brusca de Média para Mediana

```python
# Mediana do rating por faixa de tempo em 2020
median_2020 = (
    ifood_2020.groupby("faixa_tempo", observed=True)["rating"]
    .apply(lambda s: s.replace(0, np.nan).median()))

# Mediana do rating por faixa de tempo em 2021
median_2021 = (
    ifood_2021.groupby("faixa_tempo", observed=True)["rating"]
    .apply(lambda s: s.replace(0, np.nan).median()))

print("\nMedianas 2020:\n", median_2020)
print("\nMedianas 2021:\n", median_2021)
```

## 12. Correlação entre variáveis numéricas
- Selecionar apenas colunas numéricas  
- Calcular matriz de correlação com Spearman (mais robusto que Pearson)  
- Plotar mapa de calor comparando 2020 e 2021

```python
# Seleciona apenas colunas numéricas
num_2020 = ifood_2020.select_dtypes(include=["float64", "int64"])
num_2021 = ifood_2021.select_dtypes(include=["float64", "int64"])

# Matriz de correlação
corr_2020 = num_2020.corr(method="spearman")
corr_2021 = num_2021.corr(method="spearman")

# Plotar lado a lado
fig, axes = plt.subplots(1, 2, figsize=(14,6))

sns.heatmap(corr_2020, annot=True, cmap="coolwarm", vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title("Correlação das variáveis numéricas - 2020", fontsize=14, fontweight="bold")
axes[0].tick_params(axis="x", labelsize=10, rotation=45)
axes[0].tick_params(axis="y", labelsize=10, rotation=0)

sns.heatmap(corr_2021, annot=True, cmap="coolwarm", vmin=-1, vmax=1, ax=axes[1])
axes[1].set_title("Correlação das variáveis numéricas - 2021", fontsize=14, fontweight="bold")
axes[1].tick_params(axis="x", labelsize=10, rotation=45)
axes[1].tick_params(axis="y", labelsize=10, rotation=0)

plt.tight_layout()
plt.show()
```

## 13. Relação entre taxa de entrega e distância (2020 vs 2021)
- Criar faixas de distância (`bins`) para agrupar os pedidos  
- Comparar a distribuição da `delivery_fee` em cada faixa de distância  
- Plotar boxplots para 2020 e 2021 lado a lado  

```python
# Define intervalos de distância (km) e rótulos
bins   = [0, 2, 5, 10, 20]
labels = ["até 2km", "2-5km", "5-10km", "10-20km"]

# Função auxiliar para preparar dados
def prep(df):
    # Seleciona subset relevante e cria faixa de distância
    d = df.loc[df["distance"].between(0, 20), ["distance", "delivery_fee"]].copy()
    d["faixa_dist"] = pd.cut(d["distance"], bins=bins, labels=labels)
    return d

# Cria dois gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5), sharey=True, constrained_layout=True)

# Boxplot 2020
sns.boxplot(data=prep(ifood_2020), x="faixa_dist", y="delivery_fee",
            ax=axes[0], color="#FF7F0A", fliersize=2)
axes[0].set_title("Taxa x Distância (2020)", fontsize=12)
axes[0].set_xlabel("Faixa de distância", fontsize=10)
axes[0].set_ylabel("Taxa de entrega (R$)", fontsize=10)


# Boxplot 2021
sns.boxplot(data=prep(ifood_2021), x="faixa_dist", y="delivery_fee",
            ax=axes[1], color="#6A5ACD", fliersize=2)
axes[1].set_title("Taxa x Distância (2021)", fontsize=12)
axes[1].set_xlabel("Faixa de distância", fontsize=10)

plt.show()
```
